There are two types of databases
- Internal Database
- External Database

Database is a collection of tables, views and all schemas.

Table is a collection of rows and columns.

### Operations of databases

1. Create table or database
2. Insert into the table
3. Delete data from the table
4. Update the data in the table
5. Get all the data from the table

```python
# Above cell is for inserting values into table
```

```python
# inserting multiple values
yogesh = sqlite3.connect("MouriTech.db")
insert = "insert into employee(eid, ename, edept) values(102, 'Abhishek', 'Digital'), (101, 'Yogesh', 'UI/UX');"
yogesh.execute(insert)
# we need to save the data after inserting
yogesh.commit()
yogesh.close()
```

```python
# Delete entire table data
yogesh = sqlite3.connect("MouriTech.db")
delete = "delete from employee"
yogesh.execute(delete)
# we need to save the data after inserting
yogesh.commit()
yogesh.close()
```

```python
# Delete a particular data
yogesh = sqlite3.connect("MouriTech.db")
delete = "delete from employee where eid = 102"
yogesh.execute(delete)
# we need to save the data after inserting
yogesh.commit()
yogesh.close()
```

```python
# show the data of the table
yogesh = sqlite3.connect("MouriTech.db")
select = "select * from employee;"
res = yogesh.execute(select)
print(list(res))
yogesh.close()
```

```python
yogesh = sqlite3.connect("MouriTech.db")
select = "select * from employee;"
res = yogesh.execute(select)
for i in res:
    print(i[0], i[1], i[2])
yogesh.close()
```

```python
# Update the whole table
yogesh = sqlite3.connect("MouriTech.db")
update = "update employee set edept = 'python';"
yogesh.execute(update)
# we need to save the data after inserting
yogesh.commit()
yogesh.close()
```

```python
# Update particular value
yogesh = sqlite3.connect("MouriTech.db")
update = "update employee set edept = 'Digital' where eid = 102;"
yogesh.execute(update)
# we need to save the data after inserting
yogesh.commit()
yogesh.close()
```

```python
yogesh = sqlite3.connect("MouriTech.db")
update = "update employee set edept = 'Web' where eid = 101 or eid = 103 or eid = 102;"
yogesh.execute(update)
# we need to save the data after inserting
print(yogesh.total_changes) # This is for getting total number of changes in the table
yogesh.commit()
yogesh.close()
```

```python
# Adding condition(constraints) to the column of table
yogesh = sqlite3.connect("MouriTech.db")
student = "create table student(sid integer primary key, sname text, sage integer);"
yogesh.execute(student)
yogesh.close()
```

```python
yogesh = sqlite3.connect("MouriTech.db")
insert = "insert into student values(1, 'yogesh', 20)"# we cannot insert duplicate id again as 'sid' is a primary key
yogesh.execute(insert)
yogesh.commit()
yogesh.close()
```

```python
yogesh = sqlite3.connect("MouriTech.db")
select = "select * from employee order by eid;"
res = yogesh.execute(select)
for i in res:
    print(i[0], i[1], i[2])
yogesh.close()
```

```python
yogesh = sqlite3.connect("MouriTech.db")
select = "select * from employee order by eid desc;"
res = yogesh.execute(select)
for i in res:
    print(i[0], i[1], i[2])
yogesh.close()
```

```python
yogesh = sqlite3.connect("MouriTech.db")
select = "select distinct eid from employee order by eid desc;"
res = yogesh.execute(select)
for i in res:
    print(i[0])
yogesh.close()
```

```python
yogesh = sqlite3.connect("MouriTech.db")
hotel = "create table hotel(hid integer primary key autoincrement, hname text);" # system will automatically provide the id
yogesh.execute(hotel)
yogesh.close()
```

```python
yogesh = sqlite3.connect("MouriTech.db")
hotel = "insert into hotel(hname) values('Natraj');"
yogesh.execute(hotel)
yogesh.commit()
yogesh.close()
```

```python
yogesh = sqlite3.connect("MouriTech.db")
hotel = "update sqlite_sequence set seq = 12345 where name = 'hotel';" # for updating autoincreament sequence
yogesh.execute(hotel)
yogesh.commit()
yogesh.close()
```

```python
# instead of pymysql we can use mysql.connector 
yogesh = mysql.connector.connect(host="localhost", user="root", password="", database="MouriTech")
select = "select * from employee"
con = yogesh.cursor()
res = con.execute(select)
print(res)
myres = con.fetchall()
print(myres)
yogesh.close()
```

```python
# WAP to find sum of all the ids
yogesh = mysql.connector.connect(host="localhost", user="root", password="", database="MouriTech")
select = "select avg(eid) from employee"
con = yogesh.cursor()
res = con.execute(select)
print(res)
myres = con.fetchall()
print(myres)
yogesh.close()
```
