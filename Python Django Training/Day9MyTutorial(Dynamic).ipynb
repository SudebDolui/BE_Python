```python
# num = 3
# case1: 1 1 1 3 3 3 2 - 4
# case2: 1 3 1 1 1 - 3
# case3: 5 5 5 5 - 0
num = int(input())
for j in range(num):
    name = input()
    demo = name.split()
    #print(demo)
    count = 0
    for i in range(0, len(demo)-1):
        if i>0:
            if demo[i] != demo[i+1]:
                count += 1
        if i<len(demo)-1:
            if demo[i] != demo[i-1]:
                count += 1
    print(count)
```
