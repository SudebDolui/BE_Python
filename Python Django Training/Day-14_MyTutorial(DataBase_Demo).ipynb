### DataBase Demo

##### There are two types of databases
- Internal DataBase
- External Database

- Database is a collection of tables, views and all schemas.
- Table is a collection of rows and columns

#### Operations of Databases:

1. Create table or database.
2. Insert into the table.
3. Delete data from the table.
4. Update the data in the table.
5. Get all the data from the table.

SQLLite database is the default database in python

```python
import sqlite3
```

```python
db = sqlite3.connect("mouritech.db")
db.close()
```

#### Above cell is used to create database file.

```python
db = sqlite3.connect("mouritech.db")
create = "create table employee(eid integer, ename text, edept text);"
db.execute(create)
db.close()
```

#### Above cell is used to create a table 

```python
db = sqlite3.connect("mouritech.db")
insert = "insert into employee(eid, ename, edept) values(101, 'Sudeb', 'Digital')"
db.execute(insert)
# we need to save data after inserting.
db.commit()
db.close()
```

#### Above cell is used to insert value into table

```python
# Inserting multiple values

db = sqlite3.connect("mouritech.db")
insert = "insert into employee(eid, ename, edept) values(102, 'Ramya', 'Digital'), (103, 'Venu', 'UI/UX')"
db.execute(insert)
# we need to save data after inserting.
db.commit()
db.close()
```

```python
# Delete entire table data
db = sqlite3.connect("mouritech.db")
delete = "delete from employee"
db.execute(delete)
# we need to delete data after executing.
db.commit()
db.close()
```

```python
# Delete a particular data
db = sqlite3.connect("mouritech.db")
delete = "delete from employee where eid = 103"
db.execute(delete)
# we need to execute delete data command.
db.commit()
db.close()
```

```python
# Show the data of the table
db = sqlite3.connect("mouritech.db")
select = "select * from employee"
result = db.execute(select)
# we need to excute select data command and store in a variable.
print(list(result))
db.close()
```

```python
db = sqlite3.connect("mouritech.db")
select = "select * from employee"
result = db.execute(select)
for i in result:
    print(i[0], i[1], i[2])
print(list(result))
db.close()
```

```python
# show particular record from table
db = sqlite3.connect("mouritech.db")
select = "select * from employee where eid = 103;"
result = db.execute(select)
print(list(result))
db.close()
```

```python
# show record from table in order ascending
db = sqlite3.connect("mouritech.db")
select = "select * from employee order by eid;"
result = db.execute(select)
print(list(result))
db.close()
```

```python
# show record from table in order descending
db = sqlite3.connect("mouritech.db")
select = "select * from employee order by eid desc;"
result = db.execute(select)
print(list(result))
db.close()
```

```python
# show record from table using distinct(Remove duplicate)
db = sqlite3.connect("mouritech.db")
select = "select distinct eid from employee order by eid;"
result = db.execute(select)
print(list(result))
db.close()
```

```python
db = sqlite3.connect("mouritech.db")
select = "select distinct eid, rowid from employee order by eid;"
result = db.execute(select)
for i in result:
    print('\n', i[0], i[1])
db.close()
```

```python
# Update the whole table
db = sqlite3.connect("mouritech.db")
update = "update employee set edept = 'python'"
db.execute(update)
# we need to execute update data command.
db.commit()
db.close()
```

```python
db = sqlite3.connect("mouritech.db")
select = "select * from employee"
result = db.execute(select)
for i in result:
    print(i[0], i[1], i[2])
db.close()
```

```python
# Update a particular record
db = sqlite3.connect("mouritech.db")
update = "update employee set edept = 'UI/UX' where eid = 103"
db.execute(update)
# we need to execute update data command.
db.commit()
db.close()
```

```python
# Update a multpile record
db = sqlite3.connect("mouritech.db")
update = "update employee set edept = 'Front End Developer' where eid = 101 or eid = 102"
db.execute(update)
# we need to execute update data command.
db.commit()
print(db.total_changes) # This command is used for getting total number of changes that happened in the table
db.close()
```

```python
# Avoiding duplicate data
# Adding condition(Constraints) to the column of table
db = sqlite3.connect("mouritech.db")
student = "create table student(sid integer primary key, sname text, sage integer);"
db.execute(student)
db.close()
```

```python
db = sqlite3.connect("mouritech.db")
insertStudent = "insert into student values(1, 'Sudeb', 22);"
# We cannot insert duplicate record again because of primary key
db.execute(insertStudent)
db.commit()
db.close()
```

```python
db = sqlite3.connect("mouritech.db")
hotel = "create table hotel(hid integer primary key autoincrement, hname text);"
# using autoincrement system will automatically provide id
db.execute(hotel)
db.close()
```

```python
db = sqlite3.connect("mouritech.db")
insertHotel = "insert into hotel(hname) values('Natraj');"
db.execute(insertHotel)
db.commit()
db.close()
```

```python
db = sqlite3.connect("mouritech.db")
updateHotel = "update sqlite_sequence set seq=12345 where name = 'hotel';"
# for updating autoincrement sequence
db.execute(updateHotel)
db.commit()
db.close()
```
